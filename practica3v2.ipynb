```
class  Cuenta:
   def __init__(self, numero, cc, nombres, apellidos, saldo):
     self.numero = numero 
     self.cc = cc
     self.nombres = nombres
     self.apellidos = apellidos
     self.saldo = saldo
  
   def set_numero(self, numero): # modificando el valor del atributo numero
     self.numero = numero

   def get_numero(self): # retornnado el valor del atributo numero
     return self.numero

   def set_cc(self, cc): # modificando el valor del atributo cc
     self.cc = cc

   def get_cc(self): # retornnado el valor del atributo cc
     return self.cc

   def set_nombres(self, nombres): # modificando el valor del atributo nombres
     self.nombres = nombres

   def get_nombres(self): # retornnado el valor del atributo nombres
     return self.nombres

   def set_apellidos(self, apellidos): # modificando el valor del atributo apellidos
     self.apellidos = apellidos

   def get_apellidos(self): # retornnado el valor del atributo apellidos
     return self.apellidos

   def depositar(self, deposito):
         self.saldo = self.saldo + deposito

   def retirar(self, retiro):
         if retiro <= self.saldo:
           self.saldo = self.saldo - retiro
         else:
           print("Saldo insuficiente..")

   def get_saldo(self): # retornnado el valor del atributo saldo
     return self.saldo

  
   def imprimir(self):
    salida = f"CUENTA \t\t {self.numero} \n" \
             f"CEDULA \t\t {self.cc} \n" \
             f"NOMBRE \t\t {self.nombres} \n" \
             f"APELLIDO \t {self.apellidos} \n" \
             f"SALDO \t\t {self.saldo} \n"
    return salida

   

          

cuentas = []

def crearcuentas():
    cu = int(input("digite la cuenta: "))
    cc = int(input("digite cedula: "))
    no = (input("digite nombre: "))
    ap = (input("digite apellido: "))
    sa = int(input("digite saldo: "))
    nuevacuenta = Cuenta(cu,cc,no,ap,sa)
    cuentas.append(nuevacuenta)

def retiro():
  ncuenta = int(input("digite la cuenta: "))
  for x in range(0,len(cuentas)):
    if ncuenta == cuentas[x].numero:
      nretiro = int(input("digite valor retiro: "))
      if nretiro <= cuentas[x].saldo:
         cuentas[x].saldo = cuentas[x].saldo - nretiro
      else:
        print("Saldo insuficiente ... ")
      break

def deposito():
  ncuenta = int(input("digite la cuenta: "))
  for x in range(0,len(cuentas)):
    if ncuenta == cuentas[x].numero:
      ndeposito = int(input("digite valor deposito: "))
      cuentas[x].saldo = cuentas[x].saldo + ndeposito
      break

def listar():
  for x in range(0,len(cuentas)):
     print(cuentas[x].imprimir())


salir = False
while not salir:
  print("1. Crear cuentas")
  print("2. realizar retiros")
  print("3. realizar depositos")
  print("4. Imprimir datos de las cuentas")
  print("5. Salir")

  opcion = int(input("Digite la opciÃ³n "))

  if opcion == 1:
    crearcuentas()
  elif opcion == 2:
    retiro()
  elif opcion == 3:
    deposito()
  elif opcion == 4:
    listar()
  elif opcion == 5:
    salir = True
  else:
    print("la opcion debe ser 1 - 5")      

print("fin ..")


```
